```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the balanced dataset
df_balanced = pd.read_csv("E:/lightGBM model/data/Train_balanced.csv")

print("Balanced dataset loaded successfully.")
df_balanced.head()
```

```python
# Calculate the correlation matrix
corr_balanced = df_balanced.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr_balanced, cmap='coolwarm')
plt.title("Correlation Heatmap of Balanced Features")
plt.show()
```

```python
# Calculate correlation with the target variable 'labels'
corr_with_churn = df_balanced.corr(numeric_only=True)['labels'].sort_values(ascending=False)

# Plot the correlations
plt.figure(figsize=(10, 8))
sns.barplot(x=corr_with_churn.values, y=corr_with_churn.index, palette="viridis")
plt.title("Feature Correlation with Churn (Balanced Data)")
plt.xlabel("Correlation Coefficient")
plt.show()
```

```python
# Define the mapping from old names to new names
feature_mapping = {
    "feature_0": "Customer_Age", "feature_1": "Customer_Income",
    "feature_2": "Premium_Amount", "feature_3": "Tenure_Years",
    "feature_4": "Avg_Claim_Amount", "feature_5": "Num_Claims",
    "feature_6": "Policy_Cost", "feature_7": "Region",
    "feature_8": "Gender", "feature_9": "Marital_Status",
    "feature_10": "Num_Dependents", "feature_11": "Satisfaction_Score",
    "feature_12": "Loyalty_Score", "feature_13": "Years_with_Company",
    "feature_14": "Policy_Type", "feature_15": "Contact_Frequency"
}

# Apply the renaming
df_balanced.rename(columns=feature_mapping, inplace=True)

print("Features renamed for analysis.")
```

```python
# Calculate the correlation matrix
corr_balanced = df_balanced.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr_balanced, cmap='coolwarm')
plt.title("Correlation Heatmap of Balanced Features")
plt.show()
```

```python
# Calculate correlation with the target variable 'labels'
corr_with_churn = df_balanced.corr(numeric_only=True)['labels'].sort_values(ascending=False)

# Plot the correlations
plt.figure(figsize=(10, 8))
sns.barplot(x=corr_with_churn.values, y=corr_with_churn.index, palette="viridis")
plt.title("Feature Correlation with Churn (Balanced Data)")
plt.xlabel("Correlation Coefficient")
plt.show()
```
